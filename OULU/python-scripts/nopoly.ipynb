```python
import pandas as pd
import numpy as np
import logging
import scipy.io
from scipy.signal import detrend
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load necessary data
L = np.load('/data/users2/jwardell1/cholesky_decomposition.npy')
covariance_matrix = np.load('/data/users2/jwardell1/covariance_matrix.npy')

SNRs = [1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2, 2.3, 2.4, 2.5, 2.5, 2.6, 2.7, 2.8, 2.9, 3.0]
#SNRs = [2.5, 2.6, 2.7, 2.8, 2.9, 3.0]

undersampling_factor = 1
np.random.seed(42)
NOISE_SIZE = 2961 * 2
NUM_SUBS = 10
subjects = ['20150210', '20150417', '20150428', '20151110', '20151127', 
            '20150410', '20150421', '20151030', '20151117', '20151204']


logging.basicConfig(level=logging.INFO)

def pad_or_truncate(sequence, length):
    """Pad or truncate the sequence to the fixed length."""
    if len(sequence) > length:
        return sequence[:length]
    elif len(sequence) < length:
        return sequence + [0] * (length - len(sequence))
    return sequence

def safe_convert(entry):
    """Safely convert entry to a numpy array, handling NaN and non-sequences."""
    if isinstance(entry, (list, np.ndarray)):
        return np.array(entry)
    if pd.isna(entry):
        return None
    return None

def to_numpy_array(column):
    """Convert a pandas column to a numpy array, handling NaN and ensuring all entries are numpy arrays."""
    array_list = [safe_convert(entry) for entry in column]
    
    # Filter out None entries and ensure arrays have consistent shape
    valid_arrays = [arr for arr in array_list if arr is not None]
    
    if len(valid_arrays) == 0:
        raise ValueError("No valid data found in the column.")
    
    # Ensure all arrays have the same shape
    try:
        return np.vstack(valid_arrays)
    except ValueError as e:
        raise ValueError("Error stacking arrays: Ensure all sequences have the same length.") from e

def plot_tsne(X, y, title):
    """Plot t-SNE visualization."""
    tsne = TSNE(n_components=2, random_state=0)
    X_tsne = tsne.fit_transform(X)
    
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=50, alpha=0.6)
    plt.colorbar(scatter)
    plt.title(title)
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    plt.show()


for SNR in SNRs:
    num_converged = 0
    converged_subjects = []
    noises = dict()
    while num_converged < NUM_SUBS:
        for subject in subjects:
            mean = np.zeros(covariance_matrix.shape[0])
            white_noise = np.random.multivariate_normal(mean, np.eye(covariance_matrix.shape[0]), size=NOISE_SIZE)
            colored_noise = white_noise @ L.T
            noises[subject] = colored_noise.T
            num_converged += 1
            converged_subjects.append(subject)
            logging.info(f'num converged: {num_converged}')

    all_data = []

    with open('/data/users2/jwardell1/undersampling-project/OULU/txt-files/allsubs_TCs.txt', 'r') as tc_data:
        lines = tc_data.readlines()

    for i in range(0, len(lines), 2):
        subject = subjects[i // 2]
        logging.info(f'loading TC for subject {subject}')
        filepath_sr1 = lines[i].strip()
        filepath_sr2 = lines[i + 1].strip()
        try:
            sr1 = scipy.io.loadmat(filepath_sr1)['TCMax']
            sr2 = scipy.io.loadmat(filepath_sr2)['TCMax']
        except:
            continue

        if sr1.shape[0] != 53:
            sr1 = sr1.T
        if sr2.shape[0] != 53:
            sr2 = sr2.T
        if sr1.shape[1] < sr2.shape[1]:
            tr100_tc = sr2
            tr2150_tc = sr1
        else:
            tr100_tc = sr1
            tr2150_tc = sr2

        tr100_tc_zs = zscore(tr100_tc, axis=1)
        tr2150_tc_zs = zscore(tr2150_tc, axis=1)
        tr100_tc_zs_dt = detrend(tr100_tc_zs, axis=1)
        tr2150_tc_zs_dt = detrend(tr2150_tc_zs, axis=1)
        tr100_tc_zs_dt = MinMaxScaler(feature_range=(-1, 1)).fit_transform(tr100_tc_zs_dt)
        tr2150_tc_zs_dt = MinMaxScaler(feature_range=(-1, 1)).fit_transform(tr2150_tc_zs_dt)

        noise_tr100 = noises[subject][:, ::2]
        noise_tr2150 = noises[subject][:, ::33]

        tr100_tc_zs_dt_noise = tr100_tc_zs_dt + noise_tr100
        tr2150_tc_zs_dt_noise = tr2150_tc_zs_dt + noise_tr2150

        all_data.append({'Subject_ID': str(subject), 
                        'VAR_Noise': noises[subject], 
                        'TR100_Noise': noise_tr100, 
                        'TR2150_Noise': noise_tr2150, 
                        'TR100_Timecourse': tr100_tc_zs_dt, 
                        'TR2150_Timecourse': tr2150_tc_zs_dt})

    data_df = pd.DataFrame(all_data)

    xTx_tr100 = np.sum(np.square(data_df['TR100_Timecourse'].mean()))
    nTn_tr100 = np.sum(np.square(data_df['TR100_Noise'].mean()))
    scalar_tr100 = ((xTx_tr100 / nTn_tr100) ** 0.5) / (10 ** (SNR / 2)) 

    xTx_tr2150 = np.sum(np.square(data_df['TR2150_Timecourse'].mean()))
    nTn_tr2150 = np.sum(np.square(data_df['TR2150_Noise'].mean()))
    scalar_tr2150 = ((xTx_tr2150 / nTn_tr2150) ** 0.5) / (10 ** (SNR / 2)) 

    data_df['TR100_Noise'] = data_df['TR100_Noise'].multiply(scalar_tr100)
    data_df['TR2150_Noise'] = data_df['TR2150_Noise'].multiply(scalar_tr2150)

    data_df['TR100_Timecourse_Noise'] = data_df['TR100_Noise'] + data_df['TR100_Timecourse']
    data_df['TR2150_Timecourse_Noise'] = data_df['TR2150_Noise'] + data_df['TR2150_Timecourse']

    tr100_data = []
    tr2150_data = []
    add_data = []
    concat_data = []
    for subject in subjects:
        sub_row = data_df[data_df['Subject_ID'] == subject]
        logging.info(f'subject {subject}')

        tr100 = sub_row['TR100_Timecourse'].iloc[0]
        tr100_noise = sub_row['TR100_Timecourse_Noise'].iloc[0]
        
        tr2150 = sub_row['TR2150_Timecourse'].iloc[0]
        tr2150_noise = sub_row['TR2150_Timecourse_Noise'].iloc[0]

        n_regions, n_tp_tr100 = tr100.shape
        _, n_tp_tr2150 = tr2150.shape

        tr2150_window_size = 100
        tr2150_stride = 1
        n_sections = 80
        tr2150_start_ix = 0
        tr2150_end_ix = tr2150_window_size
        
        tr100_window_size = int((n_tp_tr100 / n_tp_tr2150) * tr2150_window_size)
        tr100_stride = n_tp_tr100 // n_tp_tr2150
        tr100_start_ix = 0
        tr100_end_ix = tr100_window_size

        for j in range(n_sections):
            tr100_section = tr100[:, tr100_start_ix:tr100_end_ix]
            tr100_section_noise = tr100_noise[:, tr100_start_ix:tr100_end_ix]

            tr2150_section = tr2150[:, tr2150_start_ix:tr2150_end_ix]
            tr2150_section_noise = tr2150_noise[:, tr2150_start_ix:tr2150_end_ix]

            tr100_fnc_triu = np.corrcoef(tr100_section)[np.triu_indices(n_regions)]
            tr100_noise_fnc_triu = np.corrcoef(tr100_section_noise)[np.triu_indices(n_regions)]

            tr2150_fnc_triu = np.corrcoef(tr2150_section)[np.triu_indices(n_regions)]
            tr2150_noise_fnc_triu = np.corrcoef(tr2150_section_noise)[np.triu_indices(n_regions)]

            concat_tr100_tr2150 = np.concatenate((tr100_fnc_triu, tr2150_fnc_triu))
            concat_tr100_tr2150_noise = np.concatenate((tr100_noise_fnc_triu, tr2150_noise_fnc_triu))

            add_tr100_tr2150 = tr100_fnc_triu + tr2150_fnc_triu
            add_tr100_tr2150_noise = tr100_noise_fnc_triu + tr2150_noise_fnc_triu

            tr100_data.append({'subject': subject, 
                               'TR100ms_Window': tr100_fnc_triu, 
                               'TR100ms_target': '0'})
            tr100_data.append({'subject': subject, 
                               'TR100ms_Window': tr100_noise_fnc_triu, 
                               'TR100ms_target': '1'})
            
            tr2150_data.append({'subject': subject,
                                'TR2150ms_Window': tr2150_fnc_triu, 
                                'TR2150ms_target': '0'})
            tr2150_data.append({'subject': subject,
                                'TR2150ms_Window': tr2150_noise_fnc_triu, 
                                'TR2150ms_target': '1'})
            
            concat_data.append({'subject': subject, 
                                'Concat_Window': concat_tr100_tr2150,
                                'Concat_target': '0'})
            concat_data.append({'subject': subject, 
                                'Concat_Window': concat_tr100_tr2150_noise,
                                'Concat_target': '1'})
            
            add_data.append({'subject': subject,
                             'Add_Window': add_tr100_tr2150,
                             'Add_target': '0'})
            add_data.append({'subject': subject,
                             'Add_Window': add_tr100_tr2150_noise,
                             'Add_target': '1'})

            tr100_start_ix += tr100_stride
            tr100_end_ix = tr100_end_ix + tr100_stride
                
            tr2150_start_ix += tr2150_stride
            tr2150_end_ix = tr2150_end_ix + tr2150_stride

    tr100_df = pd.DataFrame(tr100_data)
    tr2150_df = pd.DataFrame(tr2150_data)
    concat_df = pd.DataFrame(concat_data)
    add_df = pd.DataFrame(add_data)

    y_tr100 = np.array(tr100_df['TR100ms_target'])
    y_tr2150 = np.array(tr2150_df['TR2150ms_target'])
    y_concat = np.array(concat_df['Concat_target'])
    y_add = np.array(add_df['Add_target'])

    # Adding sampling rate information
    tr100_df['sampling_rate'] = 'TR100'
    tr2150_df['sampling_rate'] = 'TR2150'
    concat_df['sampling_rate'] = 'Concat'
    add_df['sampling_rate'] = 'Add'

    # Now concatenate all DataFrames
    combined_df = pd.concat([tr100_df, tr2150_df, concat_df, add_df], ignore_index=True)

    # Define the fixed length for sequences
    fixed_length = 100




    # Apply padding or truncating
    combined_df['TR100ms_Window'] = combined_df['TR100ms_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)
    combined_df['TR2150ms_Window'] = combined_df['TR2150ms_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)
    combined_df['Concat_Window'] = combined_df['Concat_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)
    combined_df['Add_Window'] = combined_df['Add_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)

    # Convert to numpy arrays
    X_tr100 = to_numpy_array(combined_df['TR100ms_Window'])
    X_tr2150 = to_numpy_array(combined_df['TR2150ms_Window'])
    X_concat = to_numpy_array(combined_df['Concat_Window'])
    X_add = to_numpy_array(combined_df['Add_Window'])



    # Define colors for sampling rates
    sampling_rate_colors = {
        'TR100': 'red',
        'TR2150': 'blue',
        'Concat': 'green',
        'Add': 'purple'
    }

    # Define markers for class labels
    class_markers = {
        '0': 'o',  # Circle
        '1': 's'   # Square
    }


    # Combine all features
    X = np.vstack([X_tr100, X_tr2150, X_concat, X_add])

    # Example target labels (replace with actual target labels)
    y = np.vstack([y_tr100, y_tr2150, y_concat, y_add])

    # Ensure X is properly shaped
    if X.ndim == 1:
        X = X.reshape(-1, 1)

    # Plot t-SNE
    plot_tsne(X, y, f't-SNE Visualization for SNR {SNR}')

    logging.info('Finished t-SNE')

```

```python
import pandas as pd
import numpy as np
import logging
import scipy.io
from scipy.signal import detrend
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load necessary data
L = np.load('/data/users2/jwardell1/cholesky_decomposition.npy')
covariance_matrix = np.load('/data/users2/jwardell1/covariance_matrix.npy')

SNRs = [1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2, 2.3, 2.4, 2.5, 2.5, 2.6, 2.7, 2.8, 2.9, 3.0]
#SNRs = [2.5, 2.6, 2.7, 2.8, 2.9, 3.0]

undersampling_factor = 1
np.random.seed(42)
NOISE_SIZE = 2961 * 2
NUM_SUBS = 10
subjects = ['20150210', '20150417', '20150428', '20151110', '20151127', 
            '20150410', '20150421', '20151030', '20151117', '20151204']

logging.basicConfig(level=logging.INFO)

def pad_or_truncate(sequence, length):
    """Pad or truncate the sequence to the fixed length."""
    if len(sequence) > length:
        return sequence[:length]
    elif len(sequence) < length:
        return sequence + [0] * (length - len(sequence))
    return sequence

def safe_convert(entry):
    """Safely convert entry to a numpy array, handling NaN and non-sequences."""
    if isinstance(entry, (list, np.ndarray)):
        return np.array(entry)
    if pd.isna(entry):
        return None
    return None

def to_numpy_array(column):
    """Convert a pandas column to a numpy array, handling NaN and ensuring all entries are numpy arrays."""
    array_list = [safe_convert(entry) for entry in column]
    
    # Filter out None entries and ensure arrays have consistent shape
    valid_arrays = [arr for arr in array_list if arr is not None]
    
    if len(valid_arrays) == 0:
        raise ValueError("No valid data found in the column.")
    
    # Ensure all arrays have the same shape
    try:
        return np.vstack(valid_arrays)
    except ValueError as e:
        raise ValueError("Error stacking arrays: Ensure all sequences have the same length.") from e

def plot_tsne(X, y, sampling_rate_labels, class_labels, title):
    """Plot t-SNE visualization with different icons for class labels and colors for sampling rates."""
    tsne = TSNE(n_components=2, random_state=0)
    X_tsne = tsne.fit_transform(X)
    
    plt.figure(figsize=(12, 10))
    
    # Define colors for sampling rates
    sampling_rate_colors = {
        'TR100': 'red',
        'TR2150': 'blue',
        'Concat': 'green',
        'Add': 'purple'
    }
    
    # Define markers for class labels
    class_markers = {
        '0': 'o',  # Circle
        '1': 's'   # Square
    }

    # Plot each combination of sampling rate and class
    for rate in sampling_rate_colors.keys():
        for cls in class_markers.keys():
            mask = (sampling_rate_labels == rate) & (class_labels == cls)
            plt.scatter(
                X_tsne[mask, 0],
                X_tsne[mask, 1],
                c=sampling_rate_colors[rate],
                marker=class_markers[cls],
                label=f'{rate} - Class {cls}',
                alpha=0.7,
                edgecolors='w'
            )

    plt.legend()
    plt.title(title)
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    plt.show()

for SNR in SNRs:
    num_converged = 0
    converged_subjects = []
    noises = dict()
    while num_converged < NUM_SUBS:
        for subject in subjects:
            mean = np.zeros(covariance_matrix.shape[0])
            white_noise = np.random.multivariate_normal(mean, np.eye(covariance_matrix.shape[0]), size=NOISE_SIZE)
            colored_noise = white_noise @ L.T
            noises[subject] = colored_noise.T
            num_converged += 1
            converged_subjects.append(subject)
            logging.info(f'num converged: {num_converged}')

    all_data = []

    with open('/data/users2/jwardell1/undersampling-project/OULU/txt-files/allsubs_TCs.txt', 'r') as tc_data:
        lines = tc_data.readlines()

    for i in range(0, len(lines), 2):
        subject = subjects[i // 2]
        logging.info(f'loading TC for subject {subject}')
        filepath_sr1 = lines[i].strip()
        filepath_sr2 = lines[i + 1].strip()
        try:
            sr1 = scipy.io.loadmat(filepath_sr1)['TCMax']
            sr2 = scipy.io.loadmat(filepath_sr2)['TCMax']
        except:
            continue

        if sr1.shape[0] != 53:
            sr1 = sr1.T
        if sr2.shape[0] != 53:
            sr2 = sr2.T
        if sr1.shape[1] < sr2.shape[1]:
            tr100_tc = sr2
            tr2150_tc = sr1
        else:
            tr100_tc = sr1
            tr2150_tc = sr2

        tr100_tc_zs = zscore(tr100_tc, axis=1)
        tr2150_tc_zs = zscore(tr2150_tc, axis=1)
        tr100_tc_zs_dt = detrend(tr100_tc_zs, axis=1)
        tr2150_tc_zs_dt = detrend(tr2150_tc_zs, axis=1)
        tr100_tc_zs_dt = MinMaxScaler(feature_range=(-1, 1)).fit_transform(tr100_tc_zs_dt)
        tr2150_tc_zs_dt = MinMaxScaler(feature_range=(-1, 1)).fit_transform(tr2150_tc_zs_dt)

        noise_tr100 = noises[subject][:, ::2]
        noise_tr2150 = noises[subject][:, ::33]

        tr100_tc_zs_dt_noise = tr100_tc_zs_dt + noise_tr100
        tr2150_tc_zs_dt_noise = tr2150_tc_zs_dt + noise_tr2150

        all_data.append({'Subject_ID': str(subject), 
                        'VAR_Noise': noises[subject], 
                        'TR100_Noise': noise_tr100, 
                        'TR2150_Noise': noise_tr2150, 
                        'TR100_Timecourse': tr100_tc_zs_dt, 
                        'TR2150_Timecourse': tr2150_tc_zs_dt})

    data_df = pd.DataFrame(all_data)

    xTx_tr100 = np.sum(np.square(data_df['TR100_Timecourse'].mean()))
    nTn_tr100 = np.sum(np.square(data_df['TR100_Noise'].mean()))
    scalar_tr100 = ((xTx_tr100 / nTn_tr100) ** 0.5) / (10 ** (SNR / 2)) 

    xTx_tr2150 = np.sum(np.square(data_df['TR2150_Timecourse'].mean()))
    nTn_tr2150 = np.sum(np.square(data_df['TR2150_Noise'].mean()))
    scalar_tr2150 = ((xTx_tr2150 / nTn_tr2150) ** 0.5) / (10 ** (SNR / 2)) 

    data_df['TR100_Noise'] = data_df['TR100_Noise'].multiply(scalar_tr100)
    data_df['TR2150_Noise'] = data_df['TR2150_Noise'].multiply(scalar_tr2150)

    data_df['TR100_Timecourse_Noise'] = data_df['TR100_Noise'] + data_df['TR100_Timecourse']
    data_df['TR2150_Timecourse_Noise'] = data_df['TR2150_Noise'] + data_df['TR2150_Timecourse']

    tr100_data = []
    tr2150_data = []
    add_data = []
    concat_data = []
    for subject in subjects:
        sub_row = data_df[data_df['Subject_ID'] == subject]
        logging.info(f'subject {subject}')

        tr100 = sub_row['TR100_Timecourse'].iloc[0]
        tr100_noise = sub_row['TR100_Timecourse_Noise'].iloc[0]
        
        tr2150 = sub_row['TR2150_Timecourse'].iloc[0]
        tr2150_noise = sub_row['TR2150_Timecourse_Noise'].iloc[0]

        n_regions, n_tp_tr100 = tr100.shape
        _, n_tp_tr2150 = tr2150.shape

        tr2150_window_size = 100
        tr2150_stride = 1
        n_sections = 80
        tr2150_start_ix = 0
        tr2150_end_ix = tr2150_window_size
        
        tr100_window_size = int((n_tp_tr100 / n_tp_tr2150) * tr2150_window_size)
        tr100_stride = n_tp_tr100 // n_tp_tr2150
        tr100_start_ix = 0
        tr100_end_ix = tr100_window_size

        for j in range(n_sections):
            tr100_section = tr100[:, tr100_start_ix:tr100_end_ix]
            tr100_section_noise = tr100_noise[:, tr100_start_ix:tr100_end_ix]

            tr2150_section = tr2150[:, tr2150_start_ix:tr2150_end_ix]
            tr2150_section_noise = tr2150_noise[:, tr2150_start_ix:tr2150_end_ix]

            tr100_fnc_triu = np.corrcoef(tr100_section)[np.triu_indices(n_regions)]
            tr100_noise_fnc_triu = np.corrcoef(tr100_section_noise)[np.triu_indices(n_regions)]

            tr2150_fnc_triu = np.corrcoef(tr2150_section)[np.triu_indices(n_regions)]
            tr2150_noise_fnc_triu = np.corrcoef(tr2150_section_noise)[np.triu_indices(n_regions)]

            concat_tr100_tr2150 = np.concatenate((tr100_fnc_triu, tr2150_fnc_triu))
            concat_tr100_tr2150_noise = np.concatenate((tr100_noise_fnc_triu, tr2150_noise_fnc_triu))

            add_tr100_tr2150 = tr100_fnc_triu + tr2150_fnc_triu
            add_tr100_tr2150_noise = tr100_noise_fnc_triu + tr2150_noise_fnc_triu

            tr100_data.append({'subject': subject, 
                               'TR100ms_Window': tr100_fnc_triu, 
                               'TR100ms_target': '0',
                               'sampling_rate': 'TR100'})
            tr100_data.append({'subject': subject, 
                               'TR100ms_Window': tr100_noise_fnc_triu, 
                               'TR100ms_target': '1',
                               'sampling_rate': 'TR100'})
            
            tr2150_data.append({'subject': subject,
                                'TR2150ms_Window': tr2150_fnc_triu, 
                                'TR2150ms_target': '0',
                                'sampling_rate': 'TR2150'})
            tr2150_data.append({'subject': subject,
                                'TR2150ms_Window': tr2150_noise_fnc_triu, 
                                'TR2150ms_target': '1',
                                'sampling_rate': 'TR2150'})
            
            concat_data.append({'subject': subject, 
                                'Concat_Window': concat_tr100_tr2150,
                                'Concat_target': '0',
                                'sampling_rate': 'Concat'})
            concat_data.append({'subject': subject, 
                                'Concat_Window': concat_tr100_tr2150_noise,
                                'Concat_target': '1',
                                'sampling_rate': 'Concat'})
            
            add_data.append({'subject': subject,
                             'Add_Window': add_tr100_tr2150,
                             'Add_target': '0',
                             'sampling_rate': 'Add'})
            add_data.append({'subject': subject,
                             'Add_Window': add_tr100_tr2150_noise,
                             'Add_target': '1',
                             'sampling_rate': 'Add'})

            tr100_start_ix += tr100_stride
            tr100_end_ix = tr100_end_ix + tr100_stride
                
            tr2150_start_ix += tr2150_stride
            tr2150_end_ix = tr2150_end_ix + tr2150_stride

    tr100_df = pd.DataFrame(tr100_data)
    tr2150_df = pd.DataFrame(tr2150_data)
    concat_df = pd.DataFrame(concat_data)
    add_df = pd.DataFrame(add_data)

    y_tr100 = np.array(tr100_df['TR100ms_target'])
    y_tr2150 = np.array(tr2150_df['TR2150ms_target'])
    y_concat = np.array(concat_df['Concat_target'])
    y_add = np.array(add_df['Add_target'])

    # Adding sampling rate information
    tr100_df['sampling_rate'] = 'TR100'
    tr2150_df['sampling_rate'] = 'TR2150'
    concat_df['sampling_rate'] = 'Concat'
    add_df['sampling_rate'] = 'Add'

    # Now concatenate all DataFrames
    combined_df = pd.concat([tr100_df, tr2150_df, concat_df, add_df], ignore_index=True)

    # Define the fixed length for sequences
    fixed_length = 100

    # Apply padding or truncating
    combined_df['TR100ms_Window'] = combined_df['TR100ms_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)
    combined_df['TR2150ms_Window'] = combined_df['TR2150ms_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)
    combined_df['Concat_Window'] = combined_df['Concat_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)
    combined_df['Add_Window'] = combined_df['Add_Window'].apply(lambda x: pad_or_truncate(x, fixed_length) if isinstance(x, (list, np.ndarray)) else np.nan)

    # Convert to numpy arrays
    X_tr100 = to_numpy_array(combined_df['TR100ms_Window'])
    X_tr2150 = to_numpy_array(combined_df['TR2150ms_Window'])
    X_concat = to_numpy_array(combined_df['Concat_Window'])
    X_add = to_numpy_array(combined_df['Add_Window'])

    # Combine all features
    X = np.vstack([X_tr100, X_tr2150, X_concat, X_add])

    # Example target labels (replace with actual target labels)
    y = np.vstack([y_tr100, y_tr2150, y_concat, y_add])

    # Ensure X is properly shaped
    if X.ndim == 1:
        X = X.reshape(-1, 1)

    # Prepare labels for t-SNE plot
    sampling_rate_labels = np.array(combined_df['sampling_rate'])
    class_labels = np.array(combined_df['TR100ms_target'].append(combined_df['TR2150ms_target']).append(combined_df['Concat_target']).append(combined_df['Add_target']))

    # Plot t-SNE
    plot_tsne(X, y, sampling_rate_labels, class_labels, f't-SNE Visualization for SNR {SNR}')

    logging.info('Finished t-SNE')
```
